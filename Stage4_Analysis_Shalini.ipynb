# Data Analysis

```python
#importing all neccessary libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
import numpy as np
import seaborn as sns 
from matplotlib import pyplot as plt
```

```python
#Reading data sets
df=pd.read_excel('E:\Dataset\Compiled HNF1A data.xlsx')
df.head()
```

```python
#preprocessing of data
df = df.drop(["PubMed ID","Sub-Population (if any)", "HGVS NM", "HGVS NP","Nucleotide Change","Protein Change","Ref","Alt","Additional Comments","Curator","Curator2","Curator3","Unnamed: 16","Unnamed: 17", "Unnamed: 18", "Unnamed: 19","Unnamed: 20","Unnamed: 21","Unnamed: 22","Unnamed: 23","Unnamed: 24","Unnamed: 25"], axis=1)
df.head()
```

```python
#Delete all rows with NaN values
df = df.dropna()
df.head()
```

```python
#Saving the processed file
df.to_csv('Data_Curated1.csv')
```

```python
#Reading data sets
data=pd.read_csv('E:\Dataset\Data_Curated1.csv')
data=data.drop('Unnamed: 0',axis=1)
data.head()
```

```python
#Enconding the data
encode = LabelEncoder()
data['Population '] = encode.fit_transform(data['Population '])
data['Population '].unique()
```

```python
#Plotting Scatter plot between Ref Frequency and Alt Frequency
plt.scatter(x=data['Frequency (Ref)'],y=data['Frequency (Alt)'],color='blue')
plt.xlabel("Ref Frequency")
plt.ylabel("Alt Frequency")
plt.title("Ref Freq vs Alt Freq")
plt.show()
```

```python
#Plotting bar graph between Ref Frequency and Alt Frequency
plt.bar(data['Frequency (Ref)'],data['Frequency (Alt)'], color ='Green')
plt.xlabel("Ref Frequency")
plt.ylabel("Alt Frequency")
plt.title("Ref Freq vs Alt Freq")
plt.show()
```

```python
#Plotting line graph between Ref Frequency and Alt Frequency
plt.plot(data['Frequency (Ref)'],data['Frequency (Alt)'], color ='orange')
plt.xlabel("Ref Frequency")
plt.ylabel("Alt Frequency")
plt.title("Ref Freq vs Alt Freq")
plt.show()
```
